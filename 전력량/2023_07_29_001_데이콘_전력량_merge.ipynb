```python
import pandas as pd
import numpy as np
import random
import os
```

```python
def seed_everything(seed):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)

seed_everything(42) # Seed 고정
```

```python
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
train_data = TabularDataset('/content/drive/MyDrive/Colab Notebooks/train_merge_scale.csv')
test_data = TabularDataset('/content/drive/MyDrive/Colab Notebooks/test_merge_scale.csv')
train_data.head()
```

```python
label = 'power'
eval_metric = 'mape'
time_limit = 3600 * 5

predictor = TabularPredictor(
    label=label, eval_metric=eval_metric
).fit(train_data, presets='best_quality', time_limit=time_limit, num_gpus=1)

predictor.leaderboard(silent = True)
```

```python
def convert_to_smape(mape): # mape 값을 smape로 변환
    smape = mape / 100  # 백분율을 소수점으로 변환하기 위해 100으로 나눕니다.
    output = 2 * smape / (1 + smape) * 100  # SMAPE 계산식을 적용하여 변환합니다.
    return output

convert_to_smape(-25810903412851.07)
```

```python
model_to_use = predictor.get_model_best()
model_pred = predictor.predict(test_data, model=model_to_use)
submission = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/submission.csv')
submission['answer'] = model_pred
submission.head()
```

```python
submission.to_csv('./병합 스케일 + autogluon_5.csv', index=False)
```

```python
train = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/train_merge_scale.csv')
train_y = train['power']
train.drop(['power'], axis = 1, inplace = True)

def SMAPE(true, pred):
    v = 2 * abs(pred - true) / (abs(pred) + abs(true))
    output = np.mean(v) * 100
    return output
```

```python
train.head()
```

```python
# import optuna
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.model_selection import train_test_split
```

```python
# 데이터 로드 및 전처리 (X_train, X_valid, y_train, y_valid)

def objective(trial):
    n_estimators = trial.suggest_int('n_estimators', 100, 1000, 100)
    max_depth = trial.suggest_int('max_depth', 10, 30, 5)
    min_samples_split = trial.suggest_int('min_samples_split', 2, 20, 2)
    bootstrap = trial.suggest_categorical('bootstrap', [True, False])

    model = ExtraTreesRegressor(
        n_estimators=n_estimators,
        min_samples_split=min_samples_split,
        max_depth=max_depth,
        bootstrap=bootstrap
    )

    model.fit(X_train, y_train)
    preds = model.predict(X_valid)
    score = SMAPE(y_valid, preds)
    return score

X_train, X_valid, y_train, y_valid = train_test_split(train, train_y, test_size = 0.2)

study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=300)
```

```python
print('Best trial:', study.best_trial.params)
print('Best score:', study.best_value)
```
